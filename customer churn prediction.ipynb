```python
data=pd.get_dummies(data=data,columns=['InternetService','Contract','PaymentMethod'])
```

```python
data.head()
```

```python
data.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
```

```python
data[['tenure',"MonthlyCharges",'TotalCharges']]=m.fit_transform(data[['tenure',"MonthlyCharges",'TotalCharges']])
```

```python
for i in data:
    print(i,data[i].unique())
```

```python
x=data.drop('Churn',axis=1)
y=data.Churn
```

```python
x.sample()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtrain[:5]
```

```python
xtrain.dtypes
```

```python
len(xtrain)
```

```python
ytrain
```

```python
ytrain.replace({'Yes':0,'No':1},inplace=True)
```

```python
ytrain
```

```python
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=100)
```

```python
ytest.replace({'Yes':0,'No':1},inplace=True)
```

```python
model.evaluate(xtest,ytest)
```

```python
y=model.predict(xtest)
```

```python
b=[]
for i in y:
    if i>0.5:
        b.append(1)
    else:
        b.append(0)
```
